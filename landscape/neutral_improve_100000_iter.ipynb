# Paysage de fitness



```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
from tqdm import tqdm

sns.set_theme()


def plot(file, title):
    fit = "score.1" if "wvcp" in file else "penalty"
    data = pd.read_csv(file)[
        [
            "%regress",
            "%neutral",
            "%improve",
            fit,
        ]
    ]#[40::]
    # plt.plot(data[fit])
    # plt.title("fitness over time")
    # plt.show()

    data = data.sort_values(by=[fit])

    print(f"{data['%regress'].max() =:.2%}")
    print(f"{data['%regress'].mean()=:.2%}")
    print(f"{data['%regress'].std() =:.2%}")
    print(f"{data['%regress'].min() =:.2%}")
    print()
    print(f"{data['%neutral'].max() =:.2%}")
    print(f"{data['%neutral'].mean()=:.2%}")
    print(f"{data['%neutral'].std() =:.2%}")
    print(f"{data['%neutral'].min() =:.2%}")
    print()
    print(f"{data['%improve'].max() =:.2%}")
    print(f"{data['%improve'].mean()=:.2%}")
    print(f"{data['%improve'].std() =:.2%}")
    print(f"{data['%improve'].min() =:.2%}")

    data.groupby(fit).mean().plot(
        kind="bar",
        y=["%neutral", "%improve"],
        # y=["%improve"],
        # y=["%regress"],
        # y=["%neutral"],
        stacked=True,
        figsize=(40, 10),
        # xlim=(1,250)
        # ylim=(1, 200),
        fontsize=20,
    )
    # plt.xticks(list(range(0, 250, 10)))
    plt.xlabel(f"Score", fontsize=30)
    plt.title(f"{title} - %neutre et %am√©liorant", fontsize=30)
    plt.show()
```

```python
# plot("run_10000_iter/gcp_DSJC500.5_random.csv", "GCP Random Walk")
plot("run_10000_iter/gcp_DSJC500.5_tabu.csv", "GCP Tabu")
# plot("run_10000_iter/wvcp_DSJC500.5_random.csv","WVCP Random Walk")
plot("run_10000_iter/wvcp_DSJC500.5_tabu.csv", "WVCP Tabu")
```
